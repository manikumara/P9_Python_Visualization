```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_used_cars = pd.read_csv('cars1.csv')
df_used_cars
```

```python
#Creating a bar plot
#Related to graph
plt.figure(    figsize=(6,4),
    facecolor='magenta',
    edgecolor='black',
    frameon=True)
#Creating a barplot using plt
plt.bar(   x= df_used_cars['Brands'],
    height = df_used_cars['Cars Listings'],
    color = 'midnightblue',
    align='center')
#X-axis related data, Names of x labels
plt.xticks (rotation = 90, fontsize = 13)
#Ttile of the graph
plt.title (label = 'Cars Listing By Brand',
        fontsize = 13,
        fontweight = 'bold',
        horizontalalignment= 'center',)
#Y axis name
plt.ylabel ('Cars Listing By Brand',
        fontsize = 13)
#Saving the plot
#plt.savefig('used cars plot.png')
```

```python
#Creating a barh plot- Everything is asame in pace of x we use y, and in pace of height, we use width
#Related to graph
plt.figure(    figsize=(9,6),
    facecolor='palegreen',
    edgecolor='black',
    frameon=True)
#Creating a barh plot using plt
plt.barh(   y= df_used_cars['Brands'],
    width = df_used_cars['Cars Listings'].sort_values(),
    color = 'midnightblue',
    align='center',)
#y-axis related data, Names of y labels
#plt.yticks (fontsize = 13)
#Ttile of the graph
plt.title (label = 'Barh graph- Cars Listing By Brand',
        fontweight = 'bold',
        horizontalalignment= 'center',)
# X axis name
plt.xlabel ('Cars Listing By Brand',
        fontsize = 13)
#Saving the plot
#plt.savefig('used cars plot.png')
```
